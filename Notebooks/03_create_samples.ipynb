```python
import pandas as pd

```

```python
data_path = "../data/processed/cleaned_dataset.csv"

df = pd.read_csv(data_path)

print("Dataset shape:", df.shape)
```

```python
df['Label'].value_counts()
```

```python
sample_50k = df.groupby('Label', group_keys=False).apply(
    lambda x: x.sample(frac=50000/len(df), random_state=42)
)

print("50K dataset shape:", sample_50k.shape)
```

```python
sample_50k.to_csv("../data/samples/dataset_50k.csv", index=False)

print("50K dataset saved successfully!")
```

```python
sample_250k = df.groupby('Label', group_keys=False).apply(
    lambda x: x.sample(frac=250000/len(df), random_state=42)
)

print("250K dataset shape:", sample_250k.shape)
```

```python
sample_250k.to_csv("../data/samples/dataset_250k.csv", index=False)

print("250K dataset saved successfully!")
```
